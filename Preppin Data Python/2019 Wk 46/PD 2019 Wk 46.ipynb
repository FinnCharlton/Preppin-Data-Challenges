```python
import pandas as pd
import numpy as np

Present = pd.read_excel("2019 PD Wk 46.xlsx",sheet_name="Present List")
Nice = pd.read_excel("2019 PD Wk 46.xlsx",sheet_name="Nice List")
Naughty = pd.read_excel("2019 PD Wk 46.xlsx",sheet_name="Naughty List")

Nice["Status"] = "Nice"
Naughty["Status"] = "Naughty"
List = pd.concat([Nice,Naughty],ignore_index=True)

Addrs = List["Address"].str.split(pat=", ",expand=True)
Addrs.loc[Addrs[2]=="Essex", "Join"] = Addrs[0]
Addrs.loc[Addrs[2]=="London", "Join"] = Addrs[1]
Addrs.loc[(Addrs[2]!="Essex") & (Addrs[2]!="London"), "Join"] = Addrs[2]
List["Join"] = Addrs["Join"]

Present["Short Name"] = Present["Name"].str[0:3]

Merge = pd.merge(Present,List,how="outer",left_on=["Short Name","Address"],right_on=["Name","Join"])

Agg = Merge.groupby("Status").agg({"Elves Build Time (min)":"sum"})
Agg["Build Time Hours"] = Agg["Elves Build Time (min)"] / 60
Agg["Build Time Hours"] = Agg["Build Time Hours"].apply(np.ceil)

Merge



```
